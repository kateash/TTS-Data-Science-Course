```python
#Top 20
top_20 = hc_data.head(20)

fig, ax = plt.subplots(figsize = (20,10))

plt.bar(top_20['Country or region'], top_20['Score'], color = 'hotpink')
length = range(len(top_20))

plt.xticks(length, top_20['Country or region'], fontsize = 20, rotation = 90)
plt.xlabel('Countries/Region', fontsize = 20)
plt.ylabel('Happiness Score', fontsize = 20)
plt.title("Top 20 Happiest Countries", fontsize = 25)
plt.show()
```

```python
hc_data = pd.read_csv('2018_Happiest_Countries.csv')
happinessrank = hc_data['Overall Happiness Rank By Country/Region']
happiestcountries = hc_data['Happiness Score By Country/Region']
gdpdata = hc_data['GDP per capita']
lifeexpectancydata = hc_data['Health Life Expectancy']
freedomdata = hc_data['Freedom to make life choices']
```

```python
#Plot graph
plt.plot(happinessrank, happiestcountries, label = 'happiness', color = 'g', marker = '*' linestyle = ':', linewidth = 3)
plt.plot(happinessrank, gdpdata, label = 'GDP per capita', color = 'm', marker = 'v', linestyle = '--', linewidth = 3)
plt.plot(happinessrank, lifeexpectancydata, label = 'Health Life Expectancy', color = 'k', marker = 'o', linestyle = 's', linewidth = 3)
plt.plot(happinessrank, freedomdata, label = 'Freedom to make life choices', color = 'hotpink', marker = 'd', linestyle = '-.', linewidth = 3)
```
