```python
#import for reassurance
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import style

#data file reassurance #Linear Regression

df = pd.read_csv('Score_GDP_Linear.csv')
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import style
from sklearn import linear_model

GDP_per_capita = df['GDP per capita'].tolist()
score = df['Score'].tolist()

x = np.array(GDP_per_capita, dtype = np.float64)
y = np.array(score, dtype = np.float64)

#Linear Regression
score_gdp = linear_model.LinearRegression()
score_gdp.fit(x.reshape(-1,1),y)
regression_line = score_gdp.predict(x.reshape(-1,1))

#Plot
style.use('seaborn')
plt.scatter(x, y, label = 'Data Points', alpha = 0.6, color = 'c', s = 70)
plt.plot(x, regression_line, label = 'Line of Best Fit', color = 'hotpink', linewidth = 3)
plt.xlabel('GDP per capita')
plt.ylabel('Happiness Score per country/region')
plt.title('Happiness Score x GDP')
plt.legend()
```

```python
#Show the coefficient and intercept
print(score_gdp.coef_)
print(score_gdp.intercept_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import style
from sklearn import linear_model

#Insert csv file
df = pd.read_csv('Score_Lifeexpectancy_Linear.csv')
df.head()
```

```python
hle = df['Healthy life expectancy'].tolist()
score = df['Score'].tolist()

x = np.array(hle, dtype = np.float64)
y = np.array(score, dtype = np.float64)

#Linear Regression
score_hle = linear_model.LinearRegression()
score_hle.fit(x.reshape(-1,1),y)
regression_line = score_hle.predict(x.reshape(-1,1))

#Plot
style.use('seaborn')
plt.scatter(x, y, label = 'Data Points', alpha = 0.6, color = 'darkviolet', s = 70)
plt.plot(x, regression_line, label = 'Line of Best Fit', color = 'dodgerblue', linewidth = 3)
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Happiness Score per country/region')
plt.title('Happiness Score x Healthy Life Expectancy')
plt.legend()
```

```python
#Show the coefficient and intercept
print(score_hle.coef_)
print(score_hle.intercept_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import style
from sklearn import linear_model

#Insert csv file
df = pd.read_csv('Score_Freedom_Linear.csv')
df.head()
```

```python
freedom = df['Freedom to make life choices'].tolist()
score = df['Score'].tolist()

x = np.array(freedom, dtype = np.float64)
y = np.array(score, dtype = np.float64)

#Linear Regression
score_freedom = linear_model.LinearRegression()
score_freedom.fit(x.reshape(-1,1),y)
regression_line = score_freedom.predict(x.reshape(-1,1))

#Plot
style.use('seaborn')
plt.scatter(x, y, label = 'Data Points', alpha = 0.6, color = 'limegreen', s = 70)
plt.plot(x, regression_line, label = 'Line of Best Fit', color = 'orange', linewidth = 3)
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness Score per country/region')
plt.title('Happiness Score x Freedom to make life choices')
plt.legend()
```

```python
#Show the coefficient and intercept
print(score_freedom.coef_)
print(score_freedom.intercept_)
```
